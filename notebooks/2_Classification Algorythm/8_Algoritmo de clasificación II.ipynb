```python
import pandas as pd
```

```python
#Leemos el Dataset
df = pd.read_csv('bank-additional-full.csv',sep = ';')
df
```

```python
df.dtypes
```

```python
# Dataframe sin la ultima columna
x=df.iloc[:, :-1].copy(deep=True)
x.head()
```

```python
#Defino mi target. 
#Es una clasificacion binaria
y = df["y"].copy(deep=True)
y
```

```python
#Transformamos el dataset en datos numericos
df=pd.get_dummies(data=df, drop_first=True)
df
```

```python
df.drop(columns='y_yes')
df.y_yes
```

```python
#Dividimos las variables independientes y la dependiente
X = df.loc[:,df.columns != 'y_yes']
y = df.loc[:,'y_yes']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Divimos nuestro Dataset en Train y Test
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Instanciamos el Modelo
dt = DecisionTreeClassifier()
```

```python
#Entremos el Modelo
dt.fit(X_train,y_train)
```

```python
y_pred_dt=dt.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
#Calculamos la Matriz de confusion
confusion_dt=confusion_matrix(y_test,y_pred_dt)
```

```python
confusion_dt
```

```python
from sklearn.metrics import accuracy_score
```

```python
#Calculamos el score de nuestro Modelo
acc_dt=accuracy_score(y_test,y_pred_dt)
```

```python
acc_dt
```

<div class="alert alert-block alert-success">
<b> Vemos que el modelo Decision Tree tiene una precision de 88.76% </b> 
</div>

```python
from sklearn.neighbors import KNeighborsClassifier # KNN
from sklearn.linear_model import LogisticRegression # Regresion Logistica
from sklearn.ensemble import RandomForestClassifier # Random Forest
```

## KNN, regresión logística y random forest

```python
#Inicializamos los modelos
knn = KNeighborsClassifier() # neighbors = 5 (DEFAULT)
lgr = LogisticRegression() # (DEFAULT)
rfc = RandomForestClassifier() # trees = 100 (DEFAULT)
```

```python
#Entrenamos el modelo KNN
knn.fit(X_train, y_train)
```

```python
#Precedimos usando el modelo KNN entrenado
y_pred_knn = knn.predict(X_test)
```

```python
#Calculamos la Matriz de confusion
confusion_knn = confusion_matrix(y_test,y_pred_knn)
```

```python
#Calculamos el score de nuestro Modelo
acc_knn=accuracy_score(y_test,y_pred_knn)
acc_knn
```

<div class="alert alert-block alert-success">
<b> Vemos que el modelo KNN tiene una precision de 90.2% </b> 
</div>

```python
#Entrenamos el modelo Logistic Regression
lgr.fit(X_train, y_train)
```

```python
#Precedimos usando el modelo LGR entrenado
y_pred_lgr = lgr.predict(X_test)
```

```python
#Calculamos la Matriz de confusion
confusion_lgr = confusion_matrix(y_test,y_pred_lgr)
```

```python
#Calculamos el score de nuestro Modelo
acc_lgr = accuracy_score(y_test,y_pred_lgr)
acc_lgr
```

<div class="alert alert-block alert-success">
<b> Vemos que el modelo Logistic Regression tiene una precision de 90.6% </b> 
</div>

```python
#Entrenamos el modelo Random Forest
rfc.fit(X_train, y_train)
```

```python
#Precedimos usando el modelo Random Forest entrenado
y_pred_rfc = rfc.predict(X_test)
```

```python
#Calculamos la Matriz de confusion
confusion_rfc = confusion_matrix(y_test,y_pred_rfc)
```

```python
#Calculamos el score de nuestro Modelo
acc_rfc = accuracy_score(y_test,y_pred_rfc)
acc_rfc
```

<div class="alert alert-block alert-success">
<b> Vemos que el modelo Random Forest tiene una precision de 91.31% </b> 
</div>
