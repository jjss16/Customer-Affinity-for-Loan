```python
import numpy as np
import pandas as pd
```

```python
#Leemos el Dataset
df = pd.read_csv('../data/bank-additional-full.csv',sep = ';')
df
```

```python
#Transformamos el dataset en datos numericos
df=pd.get_dummies(data=df, drop_first=True)
df
```

```python
#Dividimos las variables independientes y la dependiente
X = df.loc[:,df.columns != 'y_yes']
y = df.loc[:,'y_yes']
```

```python
#Divimos nuestro Dataset en Train y Test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25, random_state=12)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier(n_estimators = 200, random_state = 12,)
```

```python
model_rf.fit(X_train,y_train)
```

```python
y_pred = model_rf.predict(X_test)
```

```python
#Calculamos la Matriz de confusion
from sklearn.metrics import confusion_matrix
confusion=confusion_matrix(y_test,y_pred)
confusion
```

```python
#Calculamos el score de nuestro Modelo
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
acc
```
