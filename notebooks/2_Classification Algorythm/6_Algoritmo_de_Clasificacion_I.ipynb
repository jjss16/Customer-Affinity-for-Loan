```python
import pandas as pd
```

```python
#Leemos el Dataset
df = pd.read_csv('bank-additional-full.csv',sep = ';')
df
```

```python
df.dtypes
```

```python
# Dataframe sin la ultima columna
x=df.iloc[:, :-1].copy(deep=True)
x.head()
```

```python
#Defino mi target. 
#Es una clasificacion binaria
y = df["y"].copy(deep=True)
y
```

```python
#Transformamos el dataset en datos numericos
df=pd.get_dummies(data=df, drop_first=True)
df
```

```python
df.drop(columns='y_yes')
df.y_yes
```

```python
#Dividimos las variables independientes y la dependiente
X = df.loc[:,df.columns != 'y_yes']
y = df.loc[:,'y_yes']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Divimos nuestro Dataset en Train y Test
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Instanciamos el Modelo
model=DecisionTreeClassifier()
```

```python
model
```

```python
#Entremos el Modelo
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
#Calculamos la Matriz de confusion
confusion=confusion_matrix(y_test,y_pred)
```

```python
confusion
```

```python
from sklearn.metrics import accuracy_score
```

```python
#Calculamos el score de nuestro Modelo
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```
